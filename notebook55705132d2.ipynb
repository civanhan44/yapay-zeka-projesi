```python
import difflib

# ============================================================
# GENİŞLETİLMİŞ HASTALIK VERİTABANI
# ============================================================
veri = {
    "grip": ["yüksek ateş", "yaygın vücut ağrısı", "halsizlik", "kuru öksürük"],
    "covid": ["ateş", "öksürük", "tat/koku kaybı", "nefes darlığı", "eklem ağrısı"],
    "soguk alginligi": ["burun akıntısı", "hapşırma", "hafif boğaz ağrısı", "hafif öksürük"],
    "migren": ["tek taraflı şiddetli baş ağrısı", "bulantı", "ışığa ve sese duyarlılık"],
    "faranjit": ["şiddetli boğaz ağrısı", "yutkunma zorluğu", "boğazda yanma", "ateş"],
    "gastrit": ["mide yanması", "şişkinlik", "yemek sonrası ağrı", "bulantı"],
    "sinuzit": ["yüzde baskı ve ağrı", "burun tıkanıklığı", "geniz akıntısı", "baş ağrısı"],
    "vertigo": ["baş dönmesi", "denge kaybı", "kulak çınlaması", "mide bulantısı"],
    "diyabet": ["aşırı susama", "sık idrara çıkma", "hızlı kilo kaybı", "bulanık görme"],
    "astim": ["nefes darlığı", "hırıltılı solunum", "göğüste sıkışma", "gece öksürüğü"],
    "alerji": ["kaşıntı", "gözlerde sulanma", "hapşırma", "ciltte döküntü"],
    "miyokardit": ["göğüs ağrısı", "nefes darlığı", "anormal kalp ritmi", "yorgunluk"]
}

# ============================================================
# SORGULAMA MOTORU
# ============================================================
print("--- Hastalık Belirtisi Sorgulama Sistemi ---")
hastalik = input("Hastalık adını girin: ").lower().strip()

# 1. Tam Eşleşme veya Benzerlik Kontrolü
benzer = difflib.get_close_matches(hastalik, veri.keys(), n=1, cutoff=0.3)

if hastalik in veri:
    # Direkt bulursa
    secilen = hastalik
    belirtiler = veri[hastalik]
elif benzer:
    # Yakın eşleşme bulursa (Yazım hatası varsa)
    secilen = benzer[0]
    print(f"\n💡 Bunu mu demek istediniz: {secilen}")
    belirtiler = veri[secilen]
else:
    # Veri tabanında hiç yoksa genel belirti ver
    secilen = f"'{hastalik}' (Veritabanında Yok)"
    belirtiler = ["Genel halsizlik", "Baş ağrısı", "Vücut yorgunluğu", "Doktora danışılması önerilir"]

# ============================================================
# SONUÇ EKRANI
# ============================================================
print(f"\n🧠 {secilen.upper()} İÇİN BELİRTİLER:")
print("-" * 30)
for b in belirtiler:
    print(f"• {b}")
print("-" * 30)
```
